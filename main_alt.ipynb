```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns

# Set style for professional-looking visualizations
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['figure.dpi'] = 150
plt.rcParams['savefig.dpi'] = 300
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.size'] = 10
plt.rcParams['axes.titlesize'] = 14
plt.rcParams['axes.labelsize'] = 11
```

```python
# Load the IMF Fossil Fuel Subsidies dataset
df = pd.read_csv('Fossil_Fuel_Subsidies.csv')

# Define year columns for easy access
year_cols = [f'F{y}' for y in range(2015, 2031)]
years = list(range(2015, 2031))

# Filter to key indicators
total_implicit_explicit = df[df['Indicator'] == 'Fossil Fuel Subsidies - Total Implicit and Explicit']
explicit_total = df[df['Indicator'] == 'Explicit Fossil Fuel Subsidies - Total']
implicit_total = df[df['Indicator'] == 'Implicit Fossil Fuel Subsidies - Total']
coal_total = df[df['Indicator'] == 'Fossil Fuel Subsidies - Total Implicit and Explicit - Coal']

# Split by unit
total_usd = total_implicit_explicit[total_implicit_explicit['Unit'] == 'USD at constant 2021 prices']
total_gdp_pct = total_implicit_explicit[total_implicit_explicit['Unit'] == 'Percent of GDP']
explicit_usd = explicit_total[explicit_total['Unit'] == 'USD at constant 2021 prices']
coal_usd = coal_total[coal_total['Unit'] == 'USD at constant 2021 prices']

print(f"Dataset loaded: {len(df)} rows, {df['Country'].nunique()} countries")
print(f"Years covered: 2015-2030 (2024-2030 are IMF projections)")
```

```python
# Helper function to get data for a specific country
def get_country_data(country_name, data=total_usd):
    """Get time series data for a specific country"""
    row = data[data['Country'] == country_name]
    if len(row) == 0:
        return None
    values = [row[f'F{y}'].values[0] for y in range(2015, 2031)]
    return pd.Series(values, index=years)

# Helper function to format billions
def billions(x):
    return f"${x/1e9:.0f}B"

# Helper function to format trillions
def trillions(x):
    return f"${x/1e12:.1f}T"

# Verify data loads correctly
us_data = get_country_data('United States')
print("US Total Subsidies (sample years):")
for y in [2015, 2019, 2022, 2023]:
    print(f"  {y}: {billions(us_data[y])}")
```

## Visualization 1: SUPPORTING "Subsidies ARE Declining"
Main visualization with truncated Y-axis, cherry-picked countries, and selective time period

```python
# VISUALIZATION 1 (updated)
fig, ax = plt.subplots(figsize=(12, 7))
sns.set_theme(style="white")

plt.rcParams.update({
    "font.family": "DejaVu Serif",
    "axes.titlesize": 18,
    "axes.labelsize": 14,
    
})

declining_countries = [
    ('Saudi Arabia', 'Saudi Arabia', '#E53935'),
    ('Germany', 'Germany', '#1E88E5'),
    ('France', 'France', '#F9A825'),
    ('Italy', 'Italy', '#43A047'),
    ('Qatar', 'Qatar', '#8E24AA'),
]

chart_years = [2022, 2023, 2024, 2025]
projection_boundary = 2023.5

for full_name, short_name, color in declining_countries:
    row = total_usd[total_usd['Country'] == full_name]
    if len(row) > 0:
        values = [row[f'F{y}'].values[0] for y in chart_years]
        base_value = values[0]
        indexed_values = [(v / base_value) * 100 for v in values]

        y2022, y2023, y2024, y2025 = indexed_values

        # Interpolate halfway between 2023 and 2024
        y_boundary = y2023 + 0.5 * (y2024 - y2023)

        # Solid line up to start of projection region
        ax.plot(
            [2022, 2023, projection_boundary],
            [y2022, y2023, y_boundary],
            color=color,
            linewidth=2.8,
            linestyle='-',
            marker='o',
            markersize=10,
            markerfacecolor='white',
            markeredgewidth=2.8,
            markevery=[0, 1],
            label=short_name
        )

        # Dashed projected segment starts exactly where shaded region starts
        ax.plot(
            [projection_boundary, 2024, 2025],
            [y_boundary, y2024, y2025],
            color=color,
            linewidth=2.8,
            linestyle='--'
        )

        # Mark projected points
        ax.scatter(
            [2024, 2025],
            [y2024, y2025],
            s=110,
            facecolors='white',
            edgecolors=color,
            linewidths=2.8,
            zorder=3
        )

# Baseline at 2022 = 100
ax.axhline(y=100, color='#aaaaaa', linestyle='--', linewidth=1.2, alpha=0.8, zorder=0)

# Shaded projected region
ax.axvspan(projection_boundary, 2025.5, alpha=0.08, color='gray')

# Projection label
ax.text(
    2024.5, 91.5,
    'IMF Projected Values',
    fontsize=12,
    color='#777777',
    ha='center',
    va='center',
    style='italic'
)

# Arrow pointing to the dashed y=100 baseline
ax.annotate(
    '2022 baseline',
    xy=(2024.1, 100),
    xytext=(2024.3, 103.5),
    fontsize=12,
    fontweight='bold',
    color='#555555',
    ha='center',
    arrowprops=dict(
        arrowstyle='->',
        color='#555555',
        lw=1.8
    )
)

# Axes and limits
ax.set_xlim(2021.8, 2025.2)
ax.set_ylim(50, 108)
ax.set_xticks(chart_years)
ax.set_yticks([50, 60, 70, 80, 90, 100])

ax.set_xlabel('Year', fontsize=14, fontweight='medium')
ax.set_ylabel('Fossil Fuel Subsidy (Index: 2022)', fontsize=14, fontweight='medium')

# Title
fig.suptitle(
    'Fossil Fuel Subsidies Show Consistent Decline',
    x=0.5,
    y=0.97,
    ha='center',
    fontsize=20,
    fontweight='bold',
    color='#1B5E20'
)

# Centered subtitle
fig.text(
    0.5, 0.90,
    'Selected countries show sustained reductions relative to their 2022 subsidy levels.',
    ha='center',
    va='center',
    fontsize=11.5,
    color='dimgray'
)

# Styling
ax.tick_params(axis='both', labelsize=12)
ax.grid(True, alpha=0.22)

# Remove top and right border lines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.legend(
    loc='lower left',
    fontsize=11,
    frameon=False
)

ax.text(
    0.99, 0.02,
    'Source: IMF Fossil Fuel Subsidies Data, 2024',
    transform=ax.transAxes,
    fontsize=9,
    ha='right',
    style='italic',
    color='gray'
)

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.savefig('chart1_alt.png', dpi=300, bbox_inches='tight',
            facecolor='white', edgecolor='none')
plt.show()
```

```python
# VISUALIZATION 2: Top Emitters Subsidy Growth Since 2015
# Opposes the proposition by showing fossil fuel subsidies grew from 2015 to 2022

top_countries = [
    ('China, P.R.: Mainland', 'China'),
    ('India', 'India'),
    ('Russian Federation', 'Russia'),
    ('Japan', 'Japan'),
    ('United States', 'USA')
]

countries_data = []

for full_name, short_name in top_countries:
    row = total_usd[total_usd['Country'] == full_name]
    if len(row) > 0:
        v15 = row['F2015'].values[0] / 1e9
        v22 = row['F2022'].values[0] / 1e9
        growth = ((v22 - v15) / v15) * 100

        countries_data.append({
            'name': short_name,
            'growth': growth,
            'v15': v15,
            'v22': v22
        })

# Sort by growth percentage
countries_data.sort(key=lambda x: x['growth'], reverse=True)

names = [c['name'] for c in countries_data]
growths = [c['growth'] for c in countries_data]
v15s = [c['v15'] for c in countries_data]
v22s = [c['v22'] for c in countries_data]

fig, ax = plt.subplots(figsize=(12, 7))
sns.set_theme(style="white")

plt.rcParams.update({
    "font.family": "DejaVu Serif",
    "axes.titlesize": 18,
    "axes.labelsize": 14,
    "axes.titleweight": "bold",
    "axes.labelweight": "bold"
})

y_pos = np.arange(len(names))

colors = [
    '#C62828' if g > 50 else '#E64A19' if g > 20 else '#FF8A65'
    for g in growths
]

bars = ax.barh(
    y_pos,
    growths,
    color=colors,
    edgecolor='black',
    linewidth=1.1,
    height=0.72
)

# Bar labels
for i, (bar, growth, v15, v22, name) in enumerate(zip(bars, growths, v15s, v22s, names)):
    y_center = bar.get_y() + bar.get_height() / 2

    # Percent label at the end of each bar
    ax.text(
        bar.get_width() + 1.2,
        y_center,
        f'+{growth:.0f}%',
        va='center',
        fontsize=13,
        fontweight='bold',
        color='#B71C1C'
    )

    # Dollar labels all start at x = 15
    ax.text(
        17,
        y_center,
        f'{v15:.0f}B→{v22:.0f}B',
        va='center',
        ha='center',
        fontsize=11.5,
        color='white' if growth > 20 else '#555555',
        fontweight='bold'
    )

# Figure-level title
fig.suptitle(
    'Major Emitters Increased Fossil Fuel Subsidies Since 2015',
    x=0.5,
    y=0.97,
    ha='center',
    fontsize=22,
    fontweight='bold',
    color='#B71C1C'
)

# Subtitle
fig.text(
    0.5,
    0.91,
    'Among major emitters, subsidy growth remains substantial despite public commitments to energy transition.',
    ha='center',
    va='center',
    fontsize=12,
    color='dimgray'
)

# Annotation placed above the bars
ax.annotate(
    'India and China nearly doubled\nsubsidies since 2015',
    xy=(83, 1),
    xytext=(75, 1.75),
    fontsize=12,
    fontweight='bold',
    color='#555555',
    ha='center',
    va='center',
    arrowprops=dict(
        arrowstyle='->',
        color='#555555',
        lw=1.8
    )
)

# Axes formatting
ax.set_yticks(y_pos)
ax.set_yticklabels(names, fontsize=13)

ax.set_xlabel(
    'Growth in Fossil Fuel Subsidies, 2015–2022 (%)',
    fontsize=14,
    fontweight='bold'
)

ax.set_xlim(0, 110)
ax.set_ylim(len(names) - 0.4, -0.5)

ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=13)

ax.grid(True, alpha=0.25, axis='x')
ax.grid(False, axis='y')

# Remove top and right spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Source note
ax.text(
    0.99,
    0.02,
    'Source: IMF Fossil Fuel Subsidies Data, 2024',
    transform=ax.transAxes,
    fontsize=9,
    ha='right',
    style='italic',
    color='gray'
)

plt.tight_layout(rect=[0, 0, 1, 0.97])

fig.savefig(
    'chart3_alt.png',
    dpi=300,
    bbox_inches='tight',
    facecolor='white',
    edgecolor='none'
)

plt.show()
```

## Proposition
**"Global fossil fuel subsidies are declining."**

This proposition was chosen because the IMF Fossil Fuel Subsidies dataset contains data that can genuinely support either interpretation depending on which metrics, countries, time periods, and subsidy types are highlighted.

## Visualizations Supporting "Subsidies ARE Declining" (Page 1)

### Design Decisions

#### Decision 1: Truncated Y-Axis on US Subsidies Chart
**Score: -1.5 (Deceptive)**

The Y-axis starts at $700 billion instead of $0, making the 12% decline from $860B (2019) to $757B (2022) appear visually dramatic—occupying nearly 60% of the chart height. This is a well-known deceptive technique that exaggerates the magnitude of change. While the actual decline is real, starting the axis at zero would reveal that subsidies remain at historically high levels. I considered using a broken axis indicator but omitted it to maximize the persuasive effect.

#### Decision 2: Cherry-Picked Countries for "Leading Nations"
**Score: -1 (Deceptive)**

I selected only 5 countries (US, Canada, France, Brazil, Estonia) that show declining subsidies as % of GDP, while omitting the 136 countries where subsidies increased. The label "Leading Nations" implies these are representative examples rather than carefully selected exceptions. This survivorship bias is deceptive because it suggests a global trend that doesn't exist.

#### Decision 3: Selective Time Period (2019-2023)
**Score: -1 (Deceptive)**

By starting the US line chart at 2019 (a local peak) rather than 2015, the visualization captures a decline that might simply be normal fluctuation. If I had shown 2015-2023, the audience would see that 2022's $757B is higher than 2015's $730B—a different story entirely.

#### Decision 4: Focusing on Explicit Subsidies (2022-2023 drop)
**Score: -0.5 (Slightly Deceptive)**

The supporting chart highlights how explicit subsidies dropped 41% from 2022 to 2023, but frames the 2022 spike as an "energy crisis anomaly" rather than acknowledging that 2023 levels ($782B) are still 38% higher than 2019 ($567B). This framing dismisses inconvenient context.

#### Decision 5: Positive Framing in Titles
**Score: -1 (Deceptive)**

Titles like "Progress on Climate" and "America Leads the Way" inject editorial judgment that isn't supported by the full picture. These titles prime the viewer to interpret the data positively before examining it critically.

## Visualizations Supporting "Subsidies are NOT Declining" (Page 2)

### Design Decisions

#### Decision 1: Y-Axis Starts at Zero
**Score: +2 (Fully Earnest)**

The main chart starts the Y-axis at $0, showing the full magnitude of global subsidies ($4.4T → $7.0T → $8.2T projected). This honest scale reveals that subsidies have grown 58% since 2015 and are projected to reach $8.2 trillion by 2030.

#### Decision 2: Complete Time Series (2015-2030)
**Score: +1.5 (Earnest)**

Rather than cherry-picking time periods, I show all available years including IMF projections through 2030. This gives viewers the full context to make informed judgments. The projections are clearly labeled as such.

#### Decision 3: Including Both Actual and Projected Data
**Score: 0 (Neutral)**

While including projections is transparent (they're labeled), showing alarming growth through 2030 emphasizes worst-case scenarios. The IMF projections assume current policies continue, which may or may not occur.

#### Decision 4: Dramatic Title ("Crisis")
**Score: -0.5 (Slightly Deceptive)**

Using "crisis" is editorial, though arguably data-supported given the magnitude ($7 trillion annually, growing). Still, it primes viewers toward alarm rather than neutral assessment.

#### Decision 5: Showing All Top Emitters' Growth
**Score: +1 (Earnest)**

The supporting bar chart shows growth rates for the 5 largest emitters without excluding any. It honestly shows that India (+93%), China (+85%), and Japan (+55%) have substantially increased subsidies, while the US (+4%) has been relatively flat.

**"Declining" visualizations:**
- Truncated Y-axis (makes 12% drop look like 60% visually)

**"Not Declining" visualizations:**

## Final Reflection

The process of creating these opposing visualizations revealed how malleable data can be in the hands of a motivated visualizer. What surprised me most was how easy it was to construct a compelling "declining" narrative despite the data clearly showing the opposite trend globally. By focusing on the United States, selecting a peak-to-trough time period, truncating the Y-axis, and cherry-picking countries with declining percentages, I could construct a chart that would likely persuade a casual viewer that progress is being made—when the full picture shows subsidies nearly doubling since 2015.

Creating the "earnest" visualization (subsidies increasing) also taught me that even honest choices carry persuasive weight. Starting the Y-axis at zero is considered best practice, but it also happens to make the growth look more dramatic. Using absolute dollars rather than percentage of GDP results in larger, more alarming numbers. Including projections through 2030 shows the trend continuing, but projections are inherently uncertain. Even my choice to use red coloring—associated with danger and alarm—is a persuasive choice.

I now define "ethical analysis and visualization" as the commitment to (1) not actively misleading viewers through techniques like truncated axes or cherry-picked data, (2) providing appropriate context that allows viewers to evaluate claims critically, and (3) being transparent about choices that could influence interpretation. The line between "acceptable persuasion" and "misleading" is not always clear, but I would draw it at intentional omission of contradictory evidence and techniques (like truncated axes) that exploit visual perception to exaggerate effects. A good test: would I be embarrassed if a viewer discovered what I omitted or manipulated? If yes, it's probably over the line. The visualizations supporting "declining" subsidies failed this test; the "not declining" visualizations, despite their alarming framing, presented the full data honestly.
